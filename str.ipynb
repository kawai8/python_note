# Text sequence type --- str
class str( object ='' )   
class str( object = b'' , encoding ='utf-8' , errors ='strict' )

## str.capitalize()
- return: 最初の文字を大文字にし、残りを小文字にした文字列のコピー


```python
'python'.capitalize()
```

## str.casefold()
- return: 文字列の casefold されたコピー

```python
'ß'.casefold()
```

## str.center(width[, fillchar])
- parameter: width=int, fillchar=str
- return: width の長さをもつ中央寄せされた文字列

```python
'Python'.center(10, '+')
```

## str.count(sub[, start[, end]])
- parameter: sub=str, start=int, end=int
- return: [start, end] の範囲に、部分文字列 sub が重複せず出現する回数

```python
'I AM THAT I AM'.count('AM', 1, 4)
```

## str.encode(encoding="utf-8", errors="strict")
- parameter: encoding=[standard-encodings](https://docs.python.org/ja/3.8/library/codecs.html#standard-encodings), errors=[error-handlers](https://docs.python.org/ja/3.8/library/codecs.html#error-handlers)
- return: 文字列のエンコードされたバージョンのバイト列オブジェクト

```python
'APPLE'.encode(encoding="utf-8", errors="strict")
```

## str.endswith(suffix[, start[, end]])
- parameter: suffix=string, start=int, end=int
- return: 文字列が指定された suffix で終わるなら True

```python
'And I fear nothing. For all is as the Force wills it.'.endswith("force")
```

## str.expandtabs(tabsize=8)
- parameter: tabsize=int
- return: 文字列内の全てのタブ文字が 1 つ以上のスペースで置換された文字列のコピー

```python
'p\ty\tt\th\to\tn'.expandtabs(4)
```

## str.find(sub[, start[, end]])
- parameter: sub=string, start=int, end=int
- return: 文字列のスライス s[start:end] に部分文字列 sub が含まれる場合、その最小のインデックス

```python
'small is beautiful.'.find("l",4)
```

## str.format(\*args, **kwargs)
- parameter: value1, value2
- return: それぞれの置換フィールドが対応する引数の文字列値で置換された文字列のコピー

```python
"The sum of 1 + 2 is {0}".format(1+2)
```

## str.index(sub[, start[, end]])
- parameter: sub=string, start=int, end=int
- return: 文字列のスライス s[start:end] に部分文字列 sub が含まれる場合、その最小のインデックス  
部分文字列が見つからなかったときは ValueError 

```python
'small is beautiful.'.index("l",4)
```

## str.isalnum()
- return: 文字列中の全ての文字が英数字で、かつ 1 文字以上あるなら True

## str.lower()
- retrun:  全ての大小文字の区別のある文字が小文字に変換された文字列のコピー

```python
'PYTHON'.lower()
```

## str.upper()
- retrun: 全ての大小文字の区別のある文字が大文字に変換された文字列のコピー

```python
'python'.upper()
```

## str.swapcase()
- return: 大文字が小文字に、小文字が大文字に変換された文字列のコピー

```python
'Python'.swapcase()
```
